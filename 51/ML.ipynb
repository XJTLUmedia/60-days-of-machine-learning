```python
# Load the classification data set
data = pd.read_csv('./file.csv')
# features of interest
f = ['Temperature','Humidity', 'Light', 'CO2','HumidityRatio']
classes = ['unoccupied','occupied']
# instances and target
X = data[f]
y=data.Occupancy
```

```python
v = ScatterVisualizer(x='Light',y='CO2',classes=classes,size=(800,600))
v.fit(X,y)
v.transform(X)
v.poof()
```

```python
v = RadViz(classes=classes,features=f,size=(800,600))
v.fit(X,y)
v.transform(X)
v.poof()
```

```python
# Instantiate 
v = ParallelCoordinates(
    classes=classes, features=f,normalize='standard',sample=0.1,size=(800,600)
)
v.fit(X,y)
v.transform(X)
v.poof()
```

```python
v = Rank2D(features=f,algorithm='covariance')
v.fit(X,y)
v.transform(X)
v.poof()
```

```python
v = Rank2D(features=f,algorithm='pearson')
v.fit(X,y)
v.transform(X)
v.poof()
```
